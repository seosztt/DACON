```python
import pandas as pd
train=pd.read_csv('./train.csv')
test=pd.read_csv('./test.csv')
```

```python
train.info()
```

# 첫번째 제출

```python
color=pd.get_dummies(train['type'])
```

```python
color
```

```python
train[['red','white']]=color
```

```python
train=train.drop(['type'],axis=1)
```

```python
X_train_1=train.drop(['quality','index'], axis=1)
Y_train=train['quality']
```

```python
color=pd.get_dummies(test['type'])
```

```python
test[['red','white']]=color
```

```python
test=test.drop(['type'],axis=1)
```

```python
test_1=test.drop(['index'], axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(X_train_1, Y_train)
pred=model.predict(test_1)
```

```python
pred=list(pred)
```

```python
for i in range(1000):
    pred[i]=round(pred[i])
```
